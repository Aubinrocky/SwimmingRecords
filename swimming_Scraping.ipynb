```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import random
import time
import matplotlib.pyplot as plt
headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Ubuntu Chromium/37.0.2062.94 Chrome/37.0.2062.94 Safari/537.36'}
```

```python
def scraper(dataSet, url):
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'lxml')

    rows_prov1 = soup.find_all("tr", {"class": "rankingList0"})
    rows_prov2 = soup.find_all("tr", {"class": "rankingList1"})
    rows_prov = rows_prov1 + rows_prov2

    for row_prov in rows_prov:
        # on va chercher les textes de chaque ligne :
        liste_prov = row_prov.find_all("td")
        roww = []
        for i in liste_prov:
            if i.a != None:
                roww.append(i.a.text)
            else:
                roww.append(i.text)
        # si la discipline est un relais alors la liste a liste a que 8 colonnes, formatter les lignes
        if roww[0].startswith('4 x'):
            new = roww[6].split('\xa0')
            roww[6] = new[0] + "/" + new[1] + "/" + new[2]
            roww[7] = roww[7].split('\xa0')[0]
            roww.pop(5)
            roww.pop(2)
            roww.pop(1)
        # sinon, meme chose :
        else:
            new = roww[7].split('\xa0')
            if len(new) >= 3:
                roww[7] = new[0] + "/" + new[1] + "/" + new[2]
            else:
                roww[7] = '01/' + new[0] + "/" + new[1]
            roww[8] = roww[8].split('\xa0')[0]
            roww.pop(6)
            roww.pop(3)
            roww.pop(2)
            roww.pop(1)
        # Conversion du temps en secondes :
        time_split = roww[2].split(':')
        if len(time_split) == 1:
            roww[2] = roww[2].split('.')[0]+':'+roww[2].split('.')[1]
        if len(time_split) == 2:
            roww[2] = time_split[0] + ':' + \
                time_split[1].split('.')[0]+':'+time_split[1].split('.')[1]
        # changement format mois des dates :
        date_prov = roww[3].split('/')
        if date_prov[1] == 'Jan':
            roww[3] = date_prov[0] + '/' + '01/' + date_prov[2]
        if date_prov[1] == 'Feb':
            roww[3] = date_prov[0] + '/' + '02/' + date_prov[2]
        if date_prov[1] == 'Mar':
            roww[3] = date_prov[0] + '/' + '03/' + date_prov[2]
        if date_prov[1] == 'Apr':
            roww[3] = date_prov[0] + '/' + '04/' + date_prov[2]
        if date_prov[1] == 'May':
            roww[3] = date_prov[0] + '/' + '05/' + date_prov[2]
        if date_prov[1] == 'Jun':
            roww[3] = date_prov[0] + '/' + '06/' + date_prov[2]
        if date_prov[1] == 'Jul':
            roww[3] = date_prov[0] + '/' + '07/' + date_prov[2]
        if date_prov[1] == 'Aug':
            roww[3] = date_prov[0] + '/' + '08/' + date_prov[2]
        if date_prov[1] == 'Sep':
            roww[3] = date_prov[0] + '/' + '09/' + date_prov[2]
        if date_prov[1] == 'Oct':
            roww[3] = date_prov[0] + '/' + '10/' + date_prov[2]
        if date_prov[1] == 'Nov':
            roww[3] = date_prov[0] + '/' + '11/' + date_prov[2]
        if date_prov[1] == 'Dec':
            roww[3] = date_prov[0] + '/' + '12/' + date_prov[2]

        # On rajoute le 0 devant le dd si le dd est < Ã  10 :
        if int(date_prov[0]) < 10:
            roww[3] = '0' + roww[3]
        dataSet.append(roww)
```

```python
dataSet = []
```

```python
for annee in range(2005, 2022):
    # time.sleep(12)

    # Pour les hommes :
    # Bassin 50 m :
    time.sleep(1.3)
    
    url = 'https://www.swimrankings.net/index.php?page=rankingDetail&clubId=1&gender=1&season=' + \
        str(annee) + '&course=LCM&stroke=0&agegroup=0'
    try:
      scraper(dataSet, url)
    except:
      dataSet.append(['error','error','error','error','error'])
    print(len(dataSet))
    # Pour les hommes :
    # Bassin 25 m :
    time.sleep(2.1)
    url = 'https://www.swimrankings.net/index.php?page=rankingDetail&clubId=1&gender=1&season=' + \
        str(annee) + '&course=SCM&stroke=0&agegroup=0'
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'lxml')
    try:
      scraper(dataSet, url)
    except:
      dataSet.append(['error','error','error','error','error'])
      
    print(len(dataSet))
    # Pour les femmes :
    # Bassin 50 m :
    time.sleep(0.8)
    url = 'https://www.swimrankings.net/index.php?page=rankingDetail&clubId=1&gender=2&season=' + \
        str(annee) + '&course=LCM&stroke=0&agegroup=0'
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'lxml')
    try:
      scraper(dataSet, url)
    except:
      dataSet.append(['error','error','error','error','error'])
    print(len(dataSet))
    # Pour les femmes :
    # Bassin 25 m :
    time.sleep(1.7)
    url = 'https://www.swimrankings.net/index.php?page=rankingDetail&clubId=1&gender=2&season=' + \
        str(annee) + '&course=SCM&stroke=0&agegroup=0'
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.content, 'lxml')
    try:
      scraper(dataSet, url)
    except:
      dataSet.append(['error','error','error','error','error'])
    print(len(dataSet))
    print(annee)
```
