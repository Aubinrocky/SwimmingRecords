```python
#Importation des modules
import requests
from bs4 import BeautifulSoup
import pandas as pd
import random
import time
import matplotlib.pyplot as plt
headers = {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Ubuntu Chromium/37.0.2062.94 Chrome/37.0.2062.94 Safari/537.36'}
```

```python
disciplines = ['50_mètres_nage_libre', '100_mètres_nage_libre','200_mètres_nage_libre','400_mètres_nage_libre','1500_mètres_nage_libre',
               '50_mètres_dos','100_mètres_dos','200_mètres_dos',
               '50_mètres_brasse','100_mètres_brasse','200_mètres_brasse',
               '50_mètres_papillon','100_mètres_papillon','200_mètres_papillon',
               '200_mètres_4_nages','400_mètres_4_nages']
```

```python
dataset = []
for discipline in disciplines:
  time.sleep(0.8)
  url = 'https://fr.wikipedia.org/wiki/Record_du_monde_de_natation_messieurs_du_' + discipline
  response = requests.get(url, headers=headers)
  soup = BeautifulSoup(response.content, 'lxml')
  divs = soup.find_all("div", {"class":"mw-body-content mw-content-ltr"})
  div = divs[0]
  spans = div.find_all("span", {"class":"mw-headline"})
  tables = div.find_all("table", {"class" : "wikitable sortable"})
  
  #Bassin de 50 mètres
  trs50 = tables[0].tbody.find_all("tr")
  trs50.pop(0)
  for tr in trs50:
    ligne = []
    tds = tr.find_all("td")
    for td in tds:
      ligne.append(td.getText())
    dataset.append([discipline,"Homme","50"]+ligne)
  #Bassin de 25 mètres
  trs25 = tables[1].tbody.find_all("tr")
  trs25.pop(0)
  for tr in trs25:
    ligne = []
    tds = tr.find_all("td")
    for td in tds:
      ligne.append(td.getText())
    dataset.append([discipline,"Homme","25"]+ligne)
```

```python
#Cas particulier du 800m où le tableau n'est pas le même:
time.sleep(0.8)
url = 'https://fr.wikipedia.org/wiki/Record_du_monde_de_natation_messieurs_du_800_m%C3%A8tres_nage_libre'
response = requests.get(url, headers=headers)
soup = BeautifulSoup(response.content, 'lxml')
divs = soup.find_all("div", {"class":"mw-body-content mw-content-ltr"})
div = divs[0]
spans = div.find_all("span", {"class":"mw-headline"})
tables = div.find_all("table", {"class" : "wikitable sortable"})

#Bassin de 50 mètres
trs50 = tables[0].tbody.find_all("tr")
trs50.pop(0)
for tr in trs50:
  ligne = []
  tds = tr.find_all("td")
  for td in tds:
    ligne.append(td.getText())
  if len(ligne)==6:
    dataset.append(["800_mètres_nage_libre","Homme","50", ligne[0], ligne[1], None, None, ligne[2], ligne[3],ligne[4],ligne[5]])
#Bassin de 25 mètres
trs25 = tables[1].tbody.find_all("tr")
trs25.pop(0)
for tr in trs25:
  ligne = []
  tds = tr.find_all("td")
  for td in tds:
    ligne.append(td.getText())
  dataset.append(["800_mètres_nage_libre","Homme","25"]+ligne)
```

```python
#On retire les lignes de non record (intrus du tableau)
for i in range (len(dataset)):
  if len(dataset[i])==4 or len(dataset[i])==10:
    dataset.pop(i)
```

```python
#Mise en forme de l'age, la nationalité, la competition, le lieu et la date :
for row in dataset:
  if row[6]!= None:
    row[6] = row[6].split('\n')[0]
  row[7] = row[7].split('\n')[0]
  row[8] = row[8].split('\n')[0]
  row[9] = row[9].split('\n')[0]
  row[10] = row[10].split('\n')[0]
```

```python
#Mise en forme de la date I/II
for row in dataset:
  date_prov = row[10].split(" ")
  try:
    if date_prov[1] == 'janvier':
        row[10] = date_prov[2]+'-'+'01'+'-'+ date_prov[0]
    if date_prov[1] == 'février':
        row[10] = date_prov[2]+'-'+'02'+'-'+ date_prov[0]
    if date_prov[1] == 'mars':
        row[10] = date_prov[2]+'-'+'03'+'-'+ date_prov[0]
    if date_prov[1] == 'avril':
        row[10] = date_prov[2]+'-'+'04'+'-'+ date_prov[0]
    if date_prov[1] == 'mai':
        row[10] = date_prov[2]+'-'+'05'+'-'+ date_prov[0]
    if date_prov[1] == 'juin':
        row[10] = date_prov[2]+'-'+'06'+'-'+ date_prov[0]
    if date_prov[1] == 'juillet':
        row[10] = date_prov[2]+'-'+'07'+'-'+ date_prov[0]
    if date_prov[1] == 'août':
        row[10] = date_prov[2]+'-'+'08'+'-'+ date_prov[0]
    if date_prov[1] == 'septembre':
        row[10] = date_prov[2]+'-'+'09'+'-'+ date_prov[0]
    if date_prov[1] == 'octobre':
        row[10] = date_prov[2]+'-'+'10'+'-'+ date_prov[0]
    if date_prov[1] == 'novembre':
        row[10] = date_prov[2]+'-'+'11'+'-'+ date_prov[0]
    if date_prov[1] == 'décembre':
        row[10] = date_prov[2]+'-'+'12'+'-'+ date_prov[0]
  except:
    print(row)

#Mise en forme de la date II/II
for row in dataset:
  if 'février' in row[-1]:
    row[-1]=row[-1].split('février ')[1]
    
  if '1er' in row[-1]:
    row[-1]=row[-1].split('1er')[0]+ '1'

for row in dataset:
  if '1er' in row[-1]:
    row[-1]=row[-1].split('1er')[0]+ '1'
```

```python
#Mise en forme du temps
for row in dataset:
  if len(row[3].split('\n'))!=2:
    row[3]=row[3].split('\n')[1]+'\n'

for row in dataset:
  row[3]= row[3].split('\n')[0]

for row in dataset:
  if len(row[3].split('['))!=1:
    row[3]=row[3].split('[')[0]

for row in dataset:
  row[3]=row[3].split('(')[0]

for row in dataset:
  if len(row[3].split('min'))==2:
    row[3]=row[3].split(' min ')[0]+':'+row[3].split(' min ')[1]
  
  if len(row[3].split(' s '))==2:
    row[3]=row[3].split(' s ')[0]+':'+row[3].split(' s ')[1]
```
